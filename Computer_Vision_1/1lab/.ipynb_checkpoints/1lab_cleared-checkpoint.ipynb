```python
import skimage
import skimage.io as io
import matplotlib.pyplot as plt
import numpy as np
import math
from sklearn.metrics import mean_squared_error as mse
%matplotlib inline
```

```python
def visualize(images):
    plt.figure(figsize=(20, 20))
    for i in xrange(len(images)):
        plt.subplot(1, len(images), i + 1)
        io.imshow(images[i])
    plt.show()
```

```python
def download_and_split(fname, parts=3):
    image = io.imread(fname)
    height = (image.shape[0] // parts) * parts
    return np.split(image[:height], parts)
    # images = []
    # for part in xrange(parts):
    #     images.append(image[(height * part) / parts : (height * (part + 1)) / parts])
    # return images    
```

```python
def cut_image(image):
    # h = image.shape[0]
    # w = image.shape[1]
    cut_h, cut_w = image.shape[0] // 20, image.shape[0] // 20
    return image[cut_h:-cut_h, cut_w:-cut_w]
    # return image[int(0.07 * h) : int(0.93 * h), int(0.07 * w) : int(0.93 * w)]
```

```python
def inside(x, y, image):
    return (x >= 0 and y >= 0 and x < image.shape[0] and y < image.shape[1])
```

```python
def star_dist(img, far_img, del_h, del_w):
    l_img = 0
    l_far_img = 0
    prod = 0
    for i in xrange(img.shape[0]):
        for j in xrange(img.shape[1]):
            if inside(i + del_h, j + del_w, far_img):
                l_img += img[i][j] ** 2
                l_far_img += far_img[i][j] ** 2
                prod += img[i][j] * far_img[i][j]
    return prod / math.sqrt(l_img * l_far_img)
```

```python
def mse_dist(img, far_img, del_h, del_w):
    cells = 0
    dist = 0.0
    for i in xrange(100, img.shape[0] - 100):
        for j in xrange(100, img.shape[1] - 100):
            #if inside(i - del_h, j - del_w, far_img):
                cells += 1
                dist += (img[i][j] - far_img[i - del_h][j - del_w]) ** 2
    return dist / cells
```

```python
def star_dist(img, far_img):
    return np.sum(img * far_img) / (np.sqrt(np.sum(img * img)) * np.sqrt(np.sum(far_img * far_img)))
```

```python
def get_star_shift(img, far_img):
    DELTA_H = 15
    DELTA_W = 15 
    print(img.shape)
    print('img.shape')
    print(far_img.shape)
    print('far_img.shape')
    dists = {}
    for del_h in xrange(2 * DELTA_H):
        for del_w in xrange(2 * DELTA_W):
            
            dists[(del_h, del_w)] = star_dist(img,
                    far_img[
                        del_h:del_h + img.shape[0],
                        del_w:del_w + img.shape[1]])
    
    best_dist = dists[(0, 0)]
    best_h, best_w = 0, 0
    
    for del_h in xrange(2 * DELTA_H):
        for del_w in xrange(2 * DELTA_W):
            if dists[(del_h, del_w)] > best_dist:
                best_dist = dists[(del_h, del_w)]
                best_h, best_w = del_h, del_w
    
    return far_img[best_h:best_h + img.shape[0], best_w:best_w + img.shape[1]]
```

```python
def concat_images(images):
    for i in xrange(len(images)):
        h = images[i].shape[0]
        w = images[i].shape[1]
        images[i] = np.array(images[i]).reshape(h, w, 1)
    return np.concatenate(np.array(images), axis=2)
```

```python
def align(fname):
    sub_images = download_and_split(fname, 3)
    visualize(sub_images)
    # for i in xrange(len(sub_images)):
    #     sub_images[i] = cut_image(sub_images[i])
    sub_images = [cut_image(sub_img) for sub_img in sub_images]
    visualize(sub_images)
    
    shifted_mse_images, shifted_star_images = shift_images(sub_images)
    
    visualize(shifted_mse_images)
    visualize(shifted_star_images)
    
    visualize([
            concat_images(shifted_mse_images),  
            visualize([concat_images(shifted_star_images)])
        ])
    
    
    
    
   
    
    
    
     
    
```

```python
def cut_image(image):
    h = image.shape[0]
    w = image.shape[1]
    return image[int(0.07 * h) : int(0.93 * h), int(0.07 * w) : int(0.93 * w)]
```

```python
def inside(x, y, image):
    return (x >= 0 and y >= 0 and x < image.shape[0] and y < image.shape[1])
```

```python
def star_dist(img, far_img, del_h, del_w):
    l_img = 0
    l_far_img = 0
    prod = 0
    for i in xrange(img.shape[0]):
        for j in xrange(img.shape[1]):
            if inside(i + del_h, j + del_w, far_img):
                l_img += img[i][j] ** 2
                l_far_img += far_img[i][j] ** 2
                prod += img[i][j] * far_img[i][j]
    return prod / math.sqrt(l_img * l_far_img)
```

```python
def mse_dist(img, far_img, del_h, del_w):
    cells = 0
    dist = 0.0
    for i in xrange(100, img.shape[0] - 100):
        for j in xrange(100, img.shape[1] - 100):
            #if inside(i - del_h, j - del_w, far_img):
                cells += 1
                dist += (img[i][j] - far_img[i - del_h][j - del_w]) ** 2
    return dist / cells
```

```python
def get_shift(img, far_img, dist_type='star', opt_type='max'):
    DELTA_H = 15
    DELTA_W = 3
    dists = {}
    for del_h in xrange(-DELTA_H, DELTA_H):
        for del_w in xrange(-DELTA_W, DELTA_W):
            if dist_type == 'mse':
                dists[(del_h, del_w)] = mse_dist(img, far_img, del_h, del_w)
            if dist_type == 'star':
                dists[(del_h, del_w)] = star_dist(img, far_img, del_h, del_w)
            print('del_h=' + str(del_h) + ', del_w=' + str(del_w) + ', dist=' + str(dists[(del_h, del_w)]))
                
    best_dist = dists[(0, 0)]     
    best_del_h = 0
    best_del_w = 0
        
    for del_h in xrange(-DELTA_H, DELTA_H):
        for del_w in xrange(-DELTA_W, DELTA_W):
            if opt_type=='min':
                if dists[(del_h, del_w)] < best_dist:
                    best_dist = dists[(del_h, del_w)]
                    best_del_h = del_h
                    best_del_w = del_w
            if opt_type=='max':
                if dists[(del_h, del_w)] > best_dist:
                    best_dist = dists[(del_h, del_w)]
                    best_del_h = del_h
                    best_del_w = del_w
    
    return best_del_h, best_del_w
    
    
```

```python
def shift_images(images):
    shifts = []
    shifts.append([0, 0])
    for i in xrange(1, len(images)):
        del_h, del_w = get_shift(images[0], images[i])
        shifts.append([del_h, del_w])
    
    print(shifts)
    
    max_h = - 100
    max_w = - 100
    
    for shift in shifts:
        max_h = max(max_h, shift[0])
        max_w = max(max_w, shift[1])
        
    corner_images = []
    
    for i in xrange(len(images)):
        h = max(max_h - shifts[i][0], 0)
        w = max(max_w - shifts[i][1], 0)
        corner_images.append(images[i][h:, w:])
     
    for img in corner_images:
        print (img.shape)
    
    min_h = 1000000
    min_w = 1000000
    
    for img in corner_images:
        min_h = min(min_h, img.shape[0])
        min_w = min(min_w, img.shape[1])
    
    for i in xrange(len(corner_images)):
        corner_images[i] = corner_images[i][:min_h, :min_w]
    
    return corner_images


    
```

```python
def concat_images(images):
    for i in xrange(len(images)):
        h = images[i].shape[0]
        w = images[i].shape[1]
        images[i] = np.array(images[i]).reshape(h, w, 1)
    return np.concatenate(np.array(images), axis=2)
```

```python
def align(fname):
    sub_images = download_and_split(fname, 3)
    visualize(sub_images)
    for i in xrange(len(sub_images)):
        sub_images[i] = cut_image(sub_images[i]) 
    visualize(sub_images)
    
    shifted_images = shift_images(sub_images)
    
    visualize(shifted_images)
    
    visualize([concat_images(shifted_images)])
    
    
    
   
    
    
    
     
    
```

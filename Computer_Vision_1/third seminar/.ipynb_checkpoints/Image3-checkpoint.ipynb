```python
%matplotlib inline
from skimage import io, color
import numpy as np
```

```python
def binarize(i, lim):
    return i < lim
```

```python
gray = [color.rgb2gray(i) for i in klad]
```

Бинаризуем черно-белые изображения по порогу $0.05$. Порог подобран вручную.

```python
binarized = [binarize(i, 0.05) for i in gray]
```

```python
from matplotlib import pyplot as plt
fig, ax = plt.subplots(ncols=3, figsize=(12, 8))
for ax_i, img in zip(ax, binarized):
    ax_i.imshow(img, cmap='gray')
```

Выделим связные компоненты: (объекты с меткой кратной десяти отображаются красным цветом. Чтобы объекты не сливались с фоном, при визуализации заменим метки, кратные $10$ (и не равные $0$) на $1$).

```python
from skimage import measure
labeled = [measure.label(i) for i in binarized]
```

```python
def remove10(lbl):
    def work(x):
        if x == 0 or x % 10:
            return x
        return 1
    
    return np.array([work(x) for x in lbl.reshape(lbl.shape[0] * lbl.shape[1])]).reshape(lbl.shape)

fig, ax = plt.subplots(ncols=3, figsize=(12, 8))
for ax_i, lbl in zip(ax, labeled):
    ax_i.imshow(color.label2rgb(remove10(lbl)))
```

Получим свойства компонент:

```python
tp = [measure.regionprops(i) for i in labeled]
```

Исследование показало, что периметр стрелки всегда находится в отрезке $[180, 210]$, причем никакой другой объект из представленных на картинке не дает такой периметр.

```python
def is_arrow(prop):
    return 180 <= prop.perimeter <= 210
```

Изобразим все стрелки: (Будем отображать все стрелки одним цветом, чтобы они не сливались с красным цветом фона).

```python
def remove_not_arrow(ind, x):
    if x:
        if is_arrow(tp[ind][x - 1]):
            return 1
    return 0

fig, ax = plt.subplots(ncols=3, figsize=(12, 8))
for ind, (ax_i, lbl) in enumerate(zip(ax, labeled)):
    without = np.copy(lbl)
    without = without.reshape(lbl.shape[0] * lbl.shape[1])
    without = np.array([remove_not_arrow(ind, i) for i in without])
    ax_i.imshow(color.label2rgb(without.reshape(lbl.shape)))
```

Напишем функцию, которая будет находить начальную стрелку. Для этого найдем такую точку на картинке, цвет которой близок к $(255,0,0)$, метка не равна нулю и объект, которому принадлежит точка - стрелка.

```python
def find_start(img, tpc, lbl):
    pic = img.reshape(img.shape[0] * img.shape[1], -1)
    lblr = lbl.reshape(lbl.shape[0] * lbl.shape[1])
    for ind, (i, lc) in enumerate(zip(pic, lblr)):
        if lc and is_arrow(tpc[lc - 1]) and i[0] > 250 and i[1] < 5 and i[2] < 5:
            return ind / img.shape[1], ind % img.shape[1]
```

Найдем все начальные стрелки:

```python
starts = [find_start(klad[i], tp[i], labeled[i]) for i in xrange(3)]
```

Несложно убедиться, что нашлись верные координаты:

```python
starts
```

Будем определять направление стрелки следующим образом:

Найдем центр масс стрелки $M(x_m, y_m)$.

Для центра масс $M$ найдем наиболее удаленный пиксель из этой стрелки. $A(x_a, y_a)$. Основываясь на внешнем виде стрелки, ясно, что точка $A$ будет совпадать с вершиной стрелки. В силу симметричности, центр масс $M$ будет лежать на оси симметрии стрелки. Поэтому вектор $\overrightarrow{MA}(x_a-x_m, y_a - y_m)$ будет напрввляющим вектором стрелки.

Напишем функцию, которая будет возвращать центр масс:

```python
def find_centroid(lbl, x, y):
    sum_x, sum_y, count = (0,) * 3
    for i in xrange(lbl.shape[0]):
        for j in xrange(lbl.shape[1]):
            if lbl[i][j] == lbl[x][y]:
                sum_x += i
                sum_y += j
                count += 1
    return sum_x / count, sum_y / count
```

Напишем функцию, которая будет возвращать самую удаленную точку среди всех точек того же класса, что и переданная (предполагается, что в эту функцию будет передаваться координаты центра масс стрелки):

```python
def find_farther(lbl, x, y):
    farther_x, farther_y = x, y
    sq_dist = 0.
    for i in xrange(lbl.shape[0]):
        for j in xrange(lbl.shape[1]):
            if lbl[i][j] == lbl[x][y]:
                curr_sq_dist = (x - i) ** 2 + (y - j) ** 2
                if curr_sq_dist > sq_dist:
                    sq_dist = curr_sq_dist
                    farther_x, farther_y = i, j
    return farther_x, farther_y
```

Теперь напишем функцию, которая по координате любой точки стрелки вернет координату некоторой точки объекта, на который эта стрелка указывает:

```python
def find_next_object(lbl, x, y):
    x_c, y_c = find_centroid(lbl, x, y)
    x_f, y_f = find_farther(lbl, x_c, y_c)
    delta_x = 0. + x_f - x_c
    delta_y = 0. + y_f - y_c
    from math import hypot
    vector_len = hypot(delta_x, delta_y)
    delta_x /= vector_len
    delta_y /= vector_len
    curr_x, curr_y = 0. + x_f, 0. + y_f
    while lbl[int(curr_x)][int(curr_y)] in (0, lbl[x][y]):
        curr_x += delta_x
        curr_y += delta_y
    return int(curr_x), int(curr_y)
```

Пройдем по стрелкам (идем до тех пор, пока текущий объект является стрелкой):

```python
ways = []
for i in xrange(3):
    curr = starts[i]
    way = []
    way.append(labeled[i][curr[0]][curr[1]])
    while is_arrow(tp[i][labeled[i][curr[0]][curr[1]]  - 1]):
        curr = find_next_arrow(labeled[i], *curr)
        way.append(labeled[i][curr[0]][curr[1]])
    ways.append(way)
```

Выведем полученный путь. Стрелки пути выделим синим, клад - зеленым, все остальные символы - желтым.

```python
def replace(way, x):
    if x in way[:-1]:
        return 1
    if x == way[-1]:
        return 4
    if x:
        return 2
    return 0
    
fig, ax = plt.subplots(ncols=3, figsize=(12, 8))
for ind, (ax_i, lbl) in enumerate(zip(ax, labeled)):
    without = np.copy(lbl)
    without = without.reshape(lbl.shape[0] * lbl.shape[1])
    without = np.array([replace(ways[ind], i) for i in without])
    ax_i.imshow(color.label2rgb(without.reshape(lbl.shape)))
```

```python
import numpy as np
import keras

from __future__ import print_function
from keras.datasets import cifar10
from ktraineras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Convolution2D, MaxPooling2D
from keras.utils import np_utils

import glob

import skimage.io as sio

import matplotlib.pyplot as plt

import random
```

```python
def get_sub_square(image, width, height, x, y):
    # print(image)
    delta_width = width // 2
    delta_height = height // 2
    min_x = max(0, x - delta_width)
    min_y = max(0, y - delta_height)
    max_x = min(image.shape[0] - 1, x + delta_width)
    max_y = min(image.shape[1] - 1, y + delta_height)
    if (max_x - min_x + 1) * (max_y - min_y + 1) == width * height:
        return image[min_x:max_x+1, min_y:max_y+1].copy()
    sub_rectangle =  image[min_x : max_x + 1, min_y : max_y + 1].copy()
    avg_value = np.array([np.mean(image[:,:,0]), 
                          np.mean(image[:,:,1]), 
                          np.mean(image[:,:,2])]).astype(np.uint8)
  
    sub_square = np.zeros((width, height, 3)).astype(np.uint8)
    sub_square += avg_value
  
    sub_square[min_x - x + delta_width:max_x - x + delta_width + 1,
               min_y - y + delta_height:max_y - y + delta_height + 1] = sub_rectangle
    return sub_square
```

```python
def SquareGenerator(images, marks, width=13, height=13):
    
        Xs = []
        ys = []
      
        while True:
            for image, mark in zip(images, marks):
                xy = [(x, y) for x in range(0, image.shape[0]) for y in range(0, image.shape[1])]
                random.shuffle(xy)
                for x, y in xy:
                    square = get_sub_square(image, width, height, x, y)
                    square_mark = mark[x][y]
                    Xs.append(square)
                    ys.append(square_mark)
                    if len(Xs) == 32:
                        yield np.array(Xs), np.array(ys)
                        Xs = []
                        ys = []
   
```

```python
model = Sequential()
model.add(Convolution2D(32, 3, 3, input_shape=(13, 13, 3)))
model.add(Activation('relu'))

model.add(Convolution2D(32, 3, 3))
model.add(Activation('relu'))

model.add(Convolution2D(64, 3, 3))
model.add(Activation('relu'))

model.add(Flatten())  # this converts our 3D feature maps to 1D feature vectors
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(2))
model.add(Activation('sigmoid'))

model.compile(loss='sparse_categorical_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
```

```python
TRAIN_DIR = 'pub/'
```

```python
def load_test(dir_name):
    img_names = sorted(list(glob.glob(dir_name + 'images/*.png')))
    names = [name[len(dir_name) + 7:] for name in img_names]
    mark_names = [dir_name + 'gt/' + name for name in names]
    images = [sio.imread(name) for name in img_names]
    marks = [sio.imread(name, as_grey=True) // 255 for name in mark_names]
    return images[:60], marks[:60], images[60:], marks[60:]
```

```python
train_images, train_marks, test_images, test_marks = load_test(TRAIN_DIR)
```

```python
print(len(train_images))
print(len(train_marks))
print(len(test_images))
print(len(test_marks))
```

```python
train_pairs = list(SquareGenerator(train_images, train_marks, width=13, height=13))
test_pairs = list(SquareGenerator(test_images, test_marks, width=13, height=13))
```

```python
train_x = np.vstack([pair[0] for pair in train_pairs]) #.reshape((len(pairs), w, h, channels))
train_y = np.vstack([pair[1] for pair in train_pairs]).reshape(32 * len(train_pairs))
test_x = np.vstack([pair[0] for pair in test_pairs]) #.reshape((len(pairs), w, h, channels))
test_y = np.vstack([pair[1] for pair in test_pairs]).reshape(32 * len(test_pairs))
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
```

```python
test_y = np_utils.to_categorical(test_y, 2)
train_y = np_utils.to_categorical(train_y, 2)
```

```python
model.fit_generator(SquareGenerator(train_images, train_marks, width=13, height=13),
                    samples_per_epoch=600000,
                    nb_epoch = 10,
                    verbose=1,
                    validation_data=(test_x, test_y),
                    nb_worker=1)
```
